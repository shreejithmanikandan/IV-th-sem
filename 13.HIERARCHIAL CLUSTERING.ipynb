```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.cluster.hierarchy as sc
from sklearn import datasets
from sklearn.cluster import AgglomerativeClustering
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
iris_data = pd.DataFrame(iris.data)
```

```python
iris_data
```

```python
iris_data.columns = iris.feature_names
```

```python
iris_data['flowers_type'] = iris.target
```

```python
iris_data
```

```python
iris_x = iris_data.iloc[:,[0,1,2,3]].values
```

```python
iris_y = iris_data.iloc[:,4].values
```

```python
plt.figure(figsize=(10,7))
plt.scatter(iris_x[iris_y == 0,0], iris_x[iris_y == 0,1],s = 100, c='blue',label= 'set-1')
plt.scatter(iris_x[iris_y == 1,0], iris_x[iris_y == 1,1],s = 100, c='red',label= 'set-2')
plt.scatter(iris_x[iris_y == 2,0], iris_x[iris_y == 2,1],s = 100, c='green',label= 'set-3')
plt.legend()
plt.xlabel('Sample Index')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
#PLOT DENDROGRAM
plt.figure(figsize = (20,7))
plt.title("Dendrograms")
```

```python
sc.dendrogram(sc.linkage(iris_x,method = 'ward'))
```

```python
plt.title("Dendrogram")
plt.xlabel('Sample Index')
plt.ylabel('Euclidean Distance')
```

```python
plt.figure(figsize = (20,7))
plt.title("Dendrograms")

sc.dendrogram(sc.linkage(iris_x,method = 'ward'))

plt.title("Dendrogram")
plt.xlabel('Sample Index')
plt.ylabel('Euclidean Distance')
```
