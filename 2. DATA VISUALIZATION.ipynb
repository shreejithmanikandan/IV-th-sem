```python
tip = sb.load_dataset('tips')
```

```python
tip
```

```python
tip.tail()
```

```python
tip.head()
```

# RELATIONSHIP PLOT

```python
sb.relplot(x='total_bill',y='tip',data=tip,col="time",hue="day",style="day",kind="scatter")
```

```python
sb.relplot(x='total_bill',y='tip',data=tip,col="time",hue="day",style="day",kind="line")
```

# CATEGORICAL BOX PLOT

```python
sb.catplot(x='day',y='total_bill',kind="box",data=tip)
```

# REGRESSION PLOT 

```python
sb.regplot(x='total_bill',y='tip',data=tip)
```
