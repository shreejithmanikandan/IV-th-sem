# Creating data

```python
movies = ['Interstellar','Inception','Infinity War','Dune','Harry Potter','Oppie','FordvsFerrari']
percentage = [9.5,8.5,9,8,9,9,7]
```

# importing Library

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Scatter Plot

```python
plt.scatter(movies, percentage)
```

# Line Plot

```python
plt.plot(movies, percentage, linestyle= 'solid')
```

```python
plt.plot(movies, percentage, linestyle= 'dashed')
```

```python
plt.plot(movies, percentage, linestyle= 'dashdot')
```

```python
plt.plot(movies, percentage, linestyle= 'dotted')
```

# Pie chart

```python
plt.pie(percentage, labels = movies)
```

# Bar Chat

```python
plt.bar(movies , percentage, linestyle= 'dotted')
```

# Working with inbuilt datasets

```python
sns.get_dataset_names()
```

```python
tip = sns.load_dataset('tips')
tip
```

```python
tip.tail()
```

# Relationship Plot

```python
sns.relplot(x='total_bill', y='tip', data=tip, col ="time" , hue="day" , style= "day", kind="scatter")
```

```python
sns.relplot(x='total_bill', y='tip', data=tip, col ="time" , hue="day" , style= "day", kind="line")
```

```python
sns.catplot(x='day', y='total_bill', kind='box', data=tip)
```

# Regresssio Plot

```python
sns.regplot(x='total_bill',y= 'tip',data=tip)
```
