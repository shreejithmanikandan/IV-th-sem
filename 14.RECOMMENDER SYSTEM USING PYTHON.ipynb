# IMPORT LIBRARIES

```python
import numpy as np
import pandas as pd
```

# GET THE DATA 

```python
column_names = ['user_id','item_id','rating','timestamp']
df = pd.read_csv('u.data',sep='\t',names=column_names)
```

```python
df.head()
```

```python
df.tail()
```

```python
movie_titles = pd.read_csv("Movie_Id_Titles.csv")
movie_titles.head()
```

```python
movie_titles
```

# MERGING BOTH OF THEM

```python
df = pd.merge(df,movie_titles,on = 'item_id')
df.head()
```

```python
df
```

# VISUALIZATION IMPORTS

```python
import matplotlib.pyplot as plt
import seaborn as sns  #statistics graph generation => seaborn used
sns.set_style('white')
%matplotlib inline
```

```python
df.groupby('title')['rating'].mean().sort_values(ascending = False).head()
```

```python
df.groupby('title')['rating'].count().sort_values(ascending = False).head()
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings
```

```python
ratings['num of ratings']= pd.DataFrame(df.groupby('title')['rating'].count())
ratings
```

# HISTOGRAMS

```python
plt.figure(figsize=(10,4))
ratings['num of ratings'].hist(bins=70)
```

```python
plt.figure(figsize=(10,4))
ratings['rating'].hist(bins=70,color='chocolate')
```

```python
sns.jointplot(x='rating',y='num of ratings',data=ratings,alpha=0.5)
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat
```

# MOST RATED

```python
ratings.sort_values('num of ratings',ascending=False).head(10)
```

# USER RATINGS FOR TWO MOVIES

```python
starwars_user_rating = moviemat['Star Wars (1977)']
liarliar_user_rating = moviemat['Liar Liar (1997)']
starwars_user_rating.head(50)
```

```python
liarliar_user_rating.head(50)
```

```python
similar_to_starwars = moviemat.corrwith(starwars_user_rating)
similar_to_liarliar = moviemat.corrwith(liarliar_user_rating)
```

# FOR STARWARS

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=['correlation'])
corr_starwars.dropna(inplace=True) 
corr_starwars.head()
```

```python
corr_starwars.sort_values('correlation',ascending=False).head(50)
```

```python
corr_starwars = corr_starwars.join(ratings['num of ratings'])
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['num of ratings']>100].sort_values('correlation',ascending=False).head(50)
```

# FOR LIARLIAR

```python
corr_liarliar = pd.DataFrame(similar_to_liarliar,columns=['correlation'])
corr_liarliar.dropna(inplace=True)
corr_liarliar.head()
```

```python
corr_liarliar.sort_values('correlation', ascending=False).head(50)
```

```python
corr_liarliar = corr_liarliar.join(ratings['num of ratings'])
corr_liarliar.head()
```

```python
corr_liarliar[corr_liarliar['num of ratings']>100].sort_values('correlation',ascending=False).head(50)
```
